# Fixing

```python
df['Occupation'] = df['Occupation'].replace("_______","Other")
```

```python
def Fix_age(age):     
    return int(age[:2]) if not(age[0] == '-') else int(age[1:3])
```

```python
df['Age']  = df['Age'].apply(Fix_age)
```

```python
df['Annual_Income'] = df['Annual_Income'].apply(lambda x: float(x[:-1]) if x[-1] == "_" else float(x))
```

```python
df['Num_Bank_Accounts'] = df['Num_Bank_Accounts'].apply(lambda x: abs(x))
```
